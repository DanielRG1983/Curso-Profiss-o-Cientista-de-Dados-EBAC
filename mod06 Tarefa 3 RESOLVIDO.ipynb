#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
#%matplotlib notebook

sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
#criei uma variável faixa_asfixia no dataframe pra inserir a categorização da variável APGAR5

sinasc.loc[(sinasc['APGAR5']) <= 3, 'faixa_asfixia'] = 'asfixia grave'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] <= 5),'faixa_asfixia'] = 'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] <= 7),'faixa_asfixia'] = 'asfixia leve'
sinasc.loc[sinasc['APGAR5'] >= 8, 'faixa_asfixia'] = 'normal'
sns.countplot(x='faixa_asfixia', data=sinasc)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ax = sns.boxplot(data = sinasc, y="APGAR5", x="SEXO", orient="v", hue='GESTACAO')
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela = pd.crosstab(sinasc['APGAR5'], sinasc['GESTACAO'])
tabela
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
percentual = tabela.div(tabela.sum(axis = 1), axis = 0)
percentual
```
