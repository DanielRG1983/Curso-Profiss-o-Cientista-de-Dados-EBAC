```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
%matplotlib notebook

sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(6,3))
plt.plot(sinasc['APGAR1'])
plt.figure(figsize=(6,3))
plt.plot(sinasc['APGAR5'])
plt.figure(figsize=(6,3))
plt.plot(sinasc['SEXO'].astype(str))
plt.figure(figsize=(6,3))
plt.plot(sinasc['GESTACAO'].astype(str))
plt.figure(figsize=(6,3))
plt.plot(sinasc['ESTCIVMAE'])
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.figure(figsize=(8,4))
plt.plot(sinasc['APGAR5'])
plt.plot(sinasc['CONSULTAS'])
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0,0], x='SEXO',data=sinasc).set_title('\n SEXO')
sns.countplot(ax = axes[0,1], x='GESTACAO',data=sinasc).set_title('\n GESTAÇÃO')
sns.countplot(ax = axes[1,0], x='ESTCIVMAE',data=sinasc).set_title('\n ESTCIVMAE')
sns.countplot(ax = axes[1,1], x='CONSULTAS',data=sinasc).set_title('\n CONSULTAS')
```
